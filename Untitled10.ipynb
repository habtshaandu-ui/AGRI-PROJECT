```python
"""
Maize yield prediction in Amhara (Ethiopia)  Notebook template (Debugged)

This version fixes environments where raster-processing libraries (e.g. rasterio, rioxarray, rasterstats)
are not installed. If those libraries are available the notebook will run raster -> district aggregation.
If not available, the notebook falls back to using either a user-provided features CSV OR a small
synthetic demo dataset so the rest of the ML pipeline can run for testing and debugging.

Sections included:
- Robust imports (graceful fallbacks)
- Config / file paths
- Data loading (CSV or raster aggregation if supported)
- Synthetic demo data generator (used when raster libs are missing and no features CSV provided)
- Feature engineering (already pre-aggregated in CSV or from synthetic data)
- Modeling (Random Forest + XGBoost)
- Evaluation (RMSE, MAE, R2, feature importance)
- Visualization (scatter, simple maps using geopandas if available)

Notes for users:
- If you want to run full raster aggregation, ensure rasterio, rioxarray, and rasterstats are installed.
  If you run in a sandboxed environment where you cannot install new packages, use Google Earth Engine
  to export district-level aggregates as a CSV and place the CSV at the path configured below.

"""

# ----------------------------
# Robust imports
# ----------------------------
import os
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

# ML / stats
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import xgboost as xgb

# Optional interpretability
try:
    import shap
    HAS_SHAP = True
except Exception:
    HAS_SHAP = False

# Raster libraries may not be available in the sandbox. Try to import and set a flag.
HAS_RASTER_LIBS = True
try:
    import rasterio
    import rioxarray as rxr
    from rasterstats import zonal_stats
except Exception:
    HAS_RASTER_LIBS = False

# ----------------------------
# Config / file paths (USER: update these)
# ----------------------------
DATA_DIR = Path("./data")
paths = {
    # Yield CSV: must include columns ['district_id','year','season','yield_kg_ha']
    'yield_csv': DATA_DIR / 'yield_woreda_maize_amhara.csv',
    # Woreda shapefile (must have a district_id matching yield CSV)
    'woreda_shp': DATA_DIR / 'amhara_woreda.shp',
    # Optionally provide raster directories for NDVI / rain if raster libs exist
    'ndvi_rasters_dir': DATA_DIR / 'ndvi_ts/',
    'rain_rasters_dir': DATA_DIR / 'rain_ts/',
    # Pre-aggregated features CSV (preferred if raster libs unavailable)
    'features_csv': DATA_DIR / 'district_season_features.csv',
    'output_dir': Path('./output')
}
paths['output_dir'].mkdir(exist_ok=True)

# Project params
CROP = 'maize'
REGION = 'Amhara'
GROWING_MONTHS = {'start_month': 6, 'end_month': 11}  # example: adjust to local crop calendar
DISTRICT_ID_COL = 'district_id'
YIELD_COL = 'yield_kg_ha'
YEAR_COL = 'year'
SEED = 42

np.random.seed(SEED)

# ----------------------------
# Utility functions
# ----------------------------

def load_yield_csv(path):
    if not Path(path).exists():
        raise FileNotFoundError(f"Yield CSV not found at {path}")
    df = pd.read_csv(path)
    if DISTRICT_ID_COL not in df.columns or YIELD_COL not in df.columns:
        raise AssertionError(f"Yield CSV must contain columns: {DISTRICT_ID_COL}, {YIELD_COL}")
    return df


def load_woreda_shapefile(path):
    if not Path(path).exists():
        raise FileNotFoundError(f"Woreda shapefile not found at {path}")
    gdf = gpd.read_file(path)
    if gdf.crs is None:
        gdf.set_crs(epsg=4326, inplace=True)
    else:
        gdf = gdf.to_crs(epsg=4326)
    return gdf


def list_rasters_in_dir(dirpath):
    dirpath = Path(dirpath)
    if not dirpath.exists():
        return []
    rasters = sorted([str(p) for p in dirpath.glob('*.tif')])
    return rasters


def compute_zonal_stats_for_rasters(raster_files, vector_gdf, stats=['mean'], nodata=None):
    if not HAS_RASTER_LIBS:
        raise RuntimeError('Raster libraries (rasterio / rioxarray / rasterstats) are not available in this environment.')
    records = []
    import re
    for rf in raster_files:
        fname = Path(rf).stem
        m = re.search(r"(\d{8})", fname)
        date = pd.to_datetime(m.group(1), format='%Y%m%d') if m else None
        zs = zonal_stats(vector_gdf, rf, stats=stats, nodata=nodata, geojson_out=True)
        for feat in zs:
            props = feat['properties']
            rec = {DISTRICT_ID_COL: props.get(DISTRICT_ID_COL), 'date': date}
            for s in stats:
                rec[s] = props.get(s)
            records.append(rec)
    df = pd.DataFrame.from_records(records)
    return df

# ----------------------------
# Synthetic demo data (used when raster libs / feature CSV unavailable)
# ----------------------------

def generate_synthetic_dataset(n_districts=10, years=[2017,2018,2019,2020]):
    """
    Produces a small synthetic dataset with these columns:
    district_id, year, ndvi_mean, ndvi_max, rain_sum, rain_mean, elevation, yield_kg_ha

    The synthetic yields are a function of ndvi_mean, rain_sum and some noise so models can be trained
    and tested in environments without real data or raster libs.
    """
    rows = []
    for d in range(1, n_districts+1):
        elev = np.random.uniform(1500, 3000)  # meters
        for y in years:
            ndvi_mean = np.random.uniform(0.2, 0.7)
            ndvi_max = ndvi_mean + np.random.uniform(0.05, 0.2)
            rain_sum = np.random.uniform(200, 900)  # mm during season
            rain_mean = rain_sum / 5.0
            # synthetic yield: base + coeffs * features + noise
            yield_kg_ha = 1500 + 2500*ndvi_mean + 0.8*rain_sum - 0.2*(elev-2000) + np.random.normal(0, 200)
            rows.append({
                DISTRICT_ID_COL: d,
                YEAR_COL: y,
                'ndvi_mean': ndvi_mean,
                'ndvi_max': ndvi_max,
                'rain_sum': rain_sum,
                'rain_mean': rain_mean,
                'elevation': elev,
                YIELD_COL: max(100, yield_kg_ha)
            })
    df = pd.DataFrame(rows)
    return df

# ----------------------------
# Data loading logic (robust)
# ----------------------------
print('Starting data loading...')

if Path(paths['features_csv']).exists():
    print(f"Loading pre-aggregated features from {paths['features_csv']}")
    features_df = pd.read_csv(paths['features_csv'])
    # Basic check
    if YIELD_COL not in features_df.columns:
        raise AssertionError(f"Features CSV found but missing {YIELD_COL}")

else:
    # No features CSV  attempt raster aggregation if libs available
    if HAS_RASTER_LIBS:
        print('Raster libraries available. Attempting to compute zonal stats from raster directories.')
        # sanity checks for shapefile
        if not Path(paths['woreda_shp']).exists():
            print('Woreda shapefile not found. Falling back to synthetic demo dataset.')
            features_df = generate_synthetic_dataset()
        else:
            woreda_gdf = load_woreda_shapefile(paths['woreda_shp'])
            ndvi_rasters = list_rasters_in_dir(paths['ndvi_rasters_dir'])
            rain_rasters = list_rasters_in_dir(paths['rain_rasters_dir'])

            if len(ndvi_rasters) == 0 and len(rain_rasters) == 0:
                print('No rasters found in configured directories. Falling back to synthetic demo dataset.')
                features_df = generate_synthetic_dataset()
            else:
                # Compute zonal stats (this example computes simple seasonal aggregates)
                ndvi_stats_df = pd.DataFrame()
                rain_stats_df = pd.DataFrame()
                if len(ndvi_rasters) > 0:
                    ndvi_stats_df = compute_zonal_stats_for_rasters(ndvi_rasters, woreda_gdf, stats=['mean','max'])
                if len(rain_rasters) > 0:
                    rain_stats_df = compute_zonal_stats_for_rasters(rain_rasters, woreda_gdf, stats=['sum','mean'])

                # Post-process: aggregate by year and merge
                if not ndvi_stats_df.empty:
                    ndvi_stats_df['year'] = pd.DatetimeIndex(ndvi_stats_df['date']).year
                    ndvi_stats_df['month'] = pd.DatetimeIndex(ndvi_stats_df['date']).month
                    growing = ndvi_stats_df[(ndvi_stats_df['month'] >= GROWING_MONTHS['start_month']) & (ndvi_stats_df['month'] <= GROWING_MONTHS['end_month'])]
                    ndvi_seasonal = growing.groupby([DISTRICT_ID_COL, 'year']).agg({'mean':'mean','max':'max'}).reset_index()
                    ndvi_seasonal = ndvi_seasonal.rename(columns={'mean':'ndvi_mean','max':'ndvi_max'})
                else:
                    ndvi_seasonal = pd.DataFrame()

                if not rain_stats_df.empty:
                    rain_stats_df['year'] = pd.DatetimeIndex(rain_stats_df['date']).year
                    rain_stats_df['month'] = pd.DatetimeIndex(rain_stats_df['date']).month
                    growing_r = rain_stats_df[(rain_stats_df['month'] >= GROWING_MONTHS['start_month']) & (rain_stats_df['month'] <= GROWING_MONTHS['end_month'])]
                    rain_seasonal = growing_r.groupby([DISTRICT_ID_COL, 'year']).agg({'sum':'sum','mean':'mean'}).reset_index()
                    rain_seasonal = rain_seasonal.rename(columns={'sum':'rain_sum','mean':'rain_mean'})
                else:
                    rain_seasonal = pd.DataFrame()

                # Try to load yields if available
                if Path(paths['yield_csv']).exists():
                    yield_df = load_yield_csv(paths['yield_csv'])
                    # build features by merging
                    feature_frames = [df for df in [ndvi_seasonal, rain_seasonal] if not df.empty]
                    if len(feature_frames) == 0:
                        print('No feature frames computed  falling back to synthetic dataset.')
                        features_df = generate_synthetic_dataset()
                    else:
                        features_df = yield_df[[DISTRICT_ID_COL, YEAR_COL, YIELD_COL]].copy()
                        for f in feature_frames:
                            features_df = features_df.merge(f, on=[DISTRICT_ID_COL, 'year'], how='left')
                        # basic fill
                        features_df.fillna(features_df.median(numeric_only=True), inplace=True)
                else:
                    print('Yield CSV not found. Cannot produce labeled features from rasters. Using synthetic dataset instead.')
                    features_df = generate_synthetic_dataset()
    else:
        # raster libs not available: fall back to synthetic dataset (or require user CSV)
        if Path(paths['yield_csv']).exists() and Path(paths['woreda_shp']).exists() and Path(paths['features_csv']).exists():
            # (unlikely branch) but attempt to load features
            features_df = pd.read_csv(paths['features_csv'])
        else:
            print('Raster libraries not present and no features CSV found. Generating synthetic demo dataset for testing.')
            features_df = generate_synthetic_dataset()

# Persist a copy of features for reproducibility
features_df.to_csv(paths['output_dir'] / 'features_used.csv', index=False)
print('Prepared features dataframe with shape:', features_df.shape)

# ----------------------------
# Basic cleaning & feature matrix
# ----------------------------
print('Preparing feature matrix...')
# Ensure required columns
if YIELD_COL not in features_df.columns:
    raise AssertionError(f"Features dataframe must contain a '{YIELD_COL}' column")

exclude_cols = [DISTRICT_ID_COL, YEAR_COL, 'season', YIELD_COL]
predictor_cols = [c for c in features_df.columns if c not in exclude_cols]

# If no predictors present, create simple engineered features from available columns
if len(predictor_cols) == 0:
    print('No predictor columns detected. Attempting to synthesize predictors from available data...')
    # synthesize simple predictors
    if 'ndvi_mean' in features_df.columns:
        predictor_cols = ['ndvi_mean', 'rain_sum'] if 'rain_sum' in features_df.columns else ['ndvi_mean']
    else:
        # fallback: create random predictors (for testing only)
        features_df['p1'] = np.random.normal(size=len(features_df))
        features_df['p2'] = np.random.normal(size=len(features_df))
        predictor_cols = ['p1', 'p2']

# fill missing predictor values
for c in predictor_cols:
    if features_df[c].isnull().any():
        features_df[c].fillna(features_df[c].median(), inplace=True)

X = features_df[predictor_cols]
y = features_df[YIELD_COL]

# ----------------------------
# Train-test split: temporal holdout when possible
# ----------------------------
print('Splitting data (temporal holdout if possible)...')
years = sorted(features_df[YEAR_COL].unique()) if YEAR_COL in features_df.columns else []
if len(years) >= 2:
    test_year = years[-1]
    train_idx = features_df[YEAR_COL] != test_year
    test_idx = features_df[YEAR_COL] == test_year
    X_train, y_train = X.loc[train_idx], y.loc[train_idx]
    X_test, y_test = X.loc[test_idx], y.loc[test_idx]
else:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=SEED)
    test_idx = X_test.index

print(f"Training samples: {len(X_train)}, Test samples: {len(X_test)}")

# ----------------------------
# Model training: Random Forest
# ----------------------------
print('Training Random Forest...')
rf = RandomForestRegressor(n_estimators=200, max_depth=12, n_jobs=-1, random_state=SEED)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)

rf_rmse = np.sqrt(mean_squared_error(y_test, rf_pred))
rf_mae = mean_absolute_error(y_test, rf_pred)
rf_r2 = r2_score(y_test, rf_pred)
print('RF RMSE:', round(rf_rmse,2))
print('RF MAE:', round(rf_mae,2))
print('RF R2:', round(rf_r2,3))

# ----------------------------
# Model training: XGBoost
# ----------------------------
print('Training XGBoost...')
xgbr = xgb.XGBRegressor(n_estimators=300, max_depth=6, learning_rate=0.05, random_state=SEED, n_jobs=-1)
# safe fit (catch early stopping if small dataset)
try:
    xgbr.fit(X_train, y_train, eval_set=[(X_test, y_test)], early_stopping_rounds=20, verbose=False)
except Exception:
    # fallback to simple fit
    xgbr.fit(X_train, y_train)

xgb_pred = xgbr.predict(X_test)

xgb_rmse = np.sqrt(mean_squared_error(y_test, xgb_pred))
xgb_mae = mean_absolute_error(y_test, xgb_pred)
xgb_r2 = r2_score(y_test, xgb_pred)
print('XGB RMSE:', round(xgb_rmse,2))
print('XGB MAE:', round(xgb_mae,2))
print('XGB R2:', round(xgb_r2,3))

# ----------------------------
# Compare models: scatter plot observed vs predicted
# ----------------------------
plt.figure(figsize=(6,6))
plt.scatter(y_test, rf_pred, label='RandomForest', alpha=0.8)
plt.scatter(y_test, xgb_pred, label='XGBoost', alpha=0.8)
mn, mx = min(y_test.min(), min(rf_pred.min(), xgb_pred.min())), max(y_test.max(), max(rf_pred.max(), xgb_pred.max()))
plt.plot([mn, mx], [mn, mx], 'k--')
plt.xlabel('Observed yield (kg/ha)')
plt.ylabel('Predicted yield (kg/ha)')
plt.legend()
plt.title('Observed vs Predicted yields (test set)')
plt.grid(True)
plt.savefig(paths['output_dir'] / 'obs_vs_pred_scatter.png', dpi=150)
plt.show()

# ----------------------------
# Feature importance
# ----------------------------
print('Computing feature importances...')
rf_imp = pd.Series(rf.feature_importances_, index=predictor_cols).sort_values(ascending=False)
xgb_imp = pd.Series(xgbr.feature_importances_, index=predictor_cols).sort_values(ascending=False)
print('\nTop RF features:\n', rf_imp.head(20))
print('\nTop XGB features:\n', xgb_imp.head(20))

rf_imp.to_csv(paths['output_dir'] / 'rf_feature_importance.csv')
xgb_imp.to_csv(paths['output_dir'] / 'xgb_feature_importance.csv')

# SHAP interpretation (optional)
if HAS_SHAP:
    try:
        print('Calculating SHAP values for XGBoost (may take a bit)...')
        explainer = shap.Explainer(xgbr)
        shap_values = explainer(X_test)
        shap.summary_plot(shap_values, X_test, show=False)
        plt.savefig(paths['output_dir'] / 'shap_summary.png', dpi=150)
        plt.close()
    except Exception as e:
        print('SHAP failed:', e)
else:
    print('SHAP not available in this environment (skipping SHAP plots).')

# ----------------------------
# Simple spatial visualization if shapefile available
# ----------------------------
if Path(paths['woreda_shp']).exists():
    try:
        woreda_gdf = load_woreda_shapefile(paths['woreda_shp'])
        # prepare test_rows as dataframe with predictions
        test_rows = features_df.loc[test_idx].copy()
        test_rows['rf_pred'] = rf_pred
        test_rows['xgb_pred'] = xgb_pred
        # merge on district & (if year exists) filter to test_year
        if YEAR_COL in test_rows.columns:
            display_year = test_rows[YEAR_COL].unique()[0]
            merge_df = test_rows[test_rows[YEAR_COL] == display_year][[DISTRICT_ID_COL, 'xgb_pred', YIELD_COL]]
        else:
            merge_df = test_rows[[DISTRICT_ID_COL, 'xgb_pred', YIELD_COL]]
        map_df = woreda_gdf.merge(merge_df, on=DISTRICT_ID_COL, how='left')
        fig, ax = plt.subplots(1,1, figsize=(8,10))
        map_df.plot(column='xgb_pred', ax=ax, legend=True, missing_kwds={'color':'lightgrey'})
        ax.set_title(f'Predicted maize yield  XGBoost  {REGION}')
        ax.axis('off')
        plt.savefig(paths['output_dir'] / 'pred_map_xgb.png', dpi=150)
        plt.show()
    except Exception as e:
        print('Spatial plotting failed:', e)
else:
    print('Woreda shapefile not available  skipping spatial plots.')

# ----------------------------
# Save models and artifacts
# ----------------------------
import joblib
joblib.dump(rf, paths['output_dir'] / 'rf_model.joblib')
joblib.dump(xgbr, paths['output_dir'] / 'xgb_model.joblib')
print('Saved models to output directory.')

# ----------------------------
# Final notes
# ----------------------------
print('\nRun complete. Summary:')
print(f"- raster libraries available: {HAS_RASTER_LIBS}")
print(f"- shap available: {HAS_SHAP}") # Corrected the check for SHAP availability
print('- If you want full raster aggregation, install rasterio, rioxarray, and rasterstats, or use Google Earth Engine to export aggregates.')
print('- If you want, I can now:')
print('   * Convert this template into a runnable .ipynb with the synthetic demo embedded')
print('   * Add Google Earth Engine code to produce district-level aggregates as CSV')
print('   * Add LSTM/temporal model example (if you have timeseries features)')
```

```python
# Web Dashboard for Maize Yield Prediction in Amhara (with Demo Data)

"""
This Streamlit-based dashboard allows visualization of maize yield predictions
produced by the ML pipeline. If no data is found, it automatically generates a
demo dataset (predictions.csv, feature_importance.csv) for testing.
"""

import os
import pandas as pd
import geopandas as gpd
import streamlit as st
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np

# ----------------------------
# Setup Directories
# ----------------------------
OUTPUT_DIR = './output'
os.makedirs(OUTPUT_DIR, exist_ok=True)
FEATURES_FILE = os.path.join(OUTPUT_DIR, 'features_used.csv')
PREDICTIONS_FILE = os.path.join(OUTPUT_DIR, 'predictions.csv')
FEATURE_IMPORTANCE_FILE = os.path.join(OUTPUT_DIR, 'feature_importance.csv')

# ----------------------------
# Generate Demo Data (if missing)
# ----------------------------
if not os.path.exists(PREDICTIONS_FILE):
    st.info("No predictions.csv found  creating demo dataset for testing...")
    np.random.seed(42)

    demo_woredas = ["Bahir Dar Zuria", "Fogera", "Dera", "Debre Tabor", "Lay Gayint"]
    demo_years = [2018, 2019, 2020, 2021, 2022]

    data = []
    for year in demo_years:
        for w in demo_woredas:
            actual = np.random.uniform(2.0, 5.0)
            predicted = actual + np.random.normal(0, 0.3)
            data.append({
                'year': year,
                'woreda': w,
                'actual_yield': round(actual, 2),
                'predicted_yield': round(predicted, 2)
            })

    demo_df = pd.DataFrame(data)
    demo_df.to_csv(PREDICTIONS_FILE, index=False)

    feat_imp = pd.DataFrame({
        'feature': ['rainfall_total', 'ndvi_mean', 'temperature_avg', 'soil_ph', 'elevation'],
        'importance': [0.35, 0.28, 0.18, 0.12, 0.07]
    })
    feat_imp.to_csv(FEATURE_IMPORTANCE_FILE, index=False)

    st.success("Demo dataset created successfully!")

# ----------------------------
# Load Data
# ----------------------------
PREDICTIONS_FILE = os.path.join(OUTPUT_DIR, 'predictions.csv')
pred_df = pd.read_csv(PREDICTIONS_FILE)

st.set_page_config(page_title="Maize Yield Prediction Dashboard", layout="wide")
st.title(" Maize Yield Prediction Dashboard  Amhara, Ethiopia")
st.sidebar.header("Filters")

# ----------------------------
# Filters
# ----------------------------
if 'year' in pred_df.columns:
    years = sorted(pred_df['year'].unique())
    selected_year = st.sidebar.selectbox("Select Year", years, index=len(years)-1)
    pred_df = pred_df[pred_df['year'] == selected_year]

if 'woreda' in pred_df.columns:
    selected_woredas = st.sidebar.multiselect("Select Woreda(s)", pred_df['woreda'].unique())
    if selected_woredas:
        pred_df = pred_df[pred_df['woreda'].isin(selected_woredas)]

# ----------------------------
# Display Summary Stats
# ----------------------------
st.subheader("Summary Statistics")
col1, col2, col3 = st.columns(3)
col1.metric("Records", len(pred_df))
if 'predicted_yield' in pred_df.columns:
    col2.metric("Avg Predicted Yield", f"{pred_df['predicted_yield'].mean():.2f}")
if 'actual_yield' in pred_df.columns:
    col3.metric("Avg Actual Yield", f"{pred_df['actual_yield'].mean():.2f}")

# ----------------------------
# Plot: Predicted vs Actual
# ----------------------------
if {'predicted_yield', 'actual_yield'}.issubset(pred_df.columns):
    st.subheader("Predicted vs Actual Yield")
    fig = px.scatter(pred_df, x='actual_yield', y='predicted_yield', color='woreda', trendline="ols")
    st.plotly_chart(fig, use_container_width=True)

# ----------------------------
# Plot: Feature Importance
# ----------------------------
if os.path.exists(FEATURE_IMPORTANCE_FILE):
    st.subheader("Feature Importance")
    feat_imp = pd.read_csv(FEATURE_IMPORTANCE_FILE)
    fig_imp = px.bar(feat_imp.sort_values('importance', ascending=False).head(15), x='importance', y='feature', orientation='h')
    st.plotly_chart(fig_imp, use_container_width=True)

st.success("Dashboard loaded successfully! Adjust filters in the sidebar to explore predictions.")
```

```python
# Gradio Web Interface for Maize Yield Prediction in Amhara (with Demo Data)

"""
This Gradio-based dashboard allows visualization of maize yield predictions
produced by the ML pipeline. If no data is found, it automatically generates a
demo dataset for testing.
"""

import os
import pandas as pd
import numpy as np
import gradio as gr
import plotly.express as px

# ----------------------------
# Setup Directories
# ----------------------------
OUTPUT_DIR = './output'
os.makedirs(OUTPUT_DIR, exist_ok=True)
PREDICTIONS_FILE = os.path.join(OUTPUT_DIR, 'predictions.csv')
FEATURE_IMPORTANCE_FILE = os.path.join(OUTPUT_DIR, 'feature_importance.csv')

# ----------------------------
# Generate Demo Data (if missing)
# ----------------------------
if not os.path.exists(PREDICTIONS_FILE):
    np.random.seed(42)
    demo_woredas = ["Bahir Dar Zuria", "Fogera", "Dera", "Debre Tabor", "Lay Gayint"]
    demo_years = [2018, 2019, 2020, 2021, 2022]
    data = []
    for year in demo_years:
        for w in demo_woredas:
            actual = np.random.uniform(2.0, 5.0)
            predicted = actual + np.random.normal(0, 0.3)
            data.append({'year': year, 'woreda': w, 'actual_yield': round(actual,2), 'predicted_yield': round(predicted,2)})
    demo_df = pd.DataFrame(data)
    demo_df.to_csv(PREDICTIONS_FILE, index=False)

    feat_imp = pd.DataFrame({
        'feature': ['rainfall_total', 'ndvi_mean', 'temperature_avg', 'soil_ph', 'elevation'],
        'importance': [0.35, 0.28, 0.18, 0.12, 0.07]
    })
    feat_imp.to_csv(FEATURE_IMPORTANCE_FILE, index=False)

# ----------------------------
# Load Data
# ----------------------------
pred_df = pd.read_csv(PREDICTIONS_FILE)
feat_imp_df = pd.read_csv(FEATURE_IMPORTANCE_FILE)

# ----------------------------
# Define Gradio Interface
# ----------------------------

def plot_yield(year, woredas):
    df = pred_df[pred_df['year']==year]
    if woredas:
        df = df[df['woreda'].isin(woredas)]

    fig_scatter = px.scatter(df, x='actual_yield', y='predicted_yield', color='woreda', trendline='ols', title=f'Predicted vs Actual Yield {year}')
    fig_feat = px.bar(feat_imp_df.sort_values('importance', ascending=False).head(15), x='importance', y='feature', orientation='h', title='Top Feature Importances')
    return fig_scatter, fig_feat

woreda_options = pred_df['woreda'].unique().tolist()
year_options = sorted(pred_df['year'].unique())

iface = gr.Interface(
    fn=plot_yield,
    inputs=[
        gr.Dropdown(choices=year_options, label='Select Year', value=year_options[-1]),
        gr.CheckboxGroup(choices=woreda_options, label='Select Woreda(s)')
    ],
    outputs=[
        gr.Plot(label='Predicted vs Actual Yield'),
        gr.Plot(label='Feature Importance')
    ],
    title=" Maize Yield Prediction Dashboard  Amhara, Ethiopia",
    description="Visualize maize yield predictions and feature importance interactively using Gradio."
)

# ----------------------------
# Launch Interface
# ----------------------------
iface.launch()
```
